```python
import numpy as np
import pandas as pd
import datetime as d
```

```python
df = pd.read_csv("hotel_bookings.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['children'] = df['children'].fillna(0)
```

```python
df.dropna(subset=['country'], inplace=True)
```

```python
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)
df.isnull().sum()
```

```python
df['children'] = df['children'].astype("int64")
df['agent'] = df['agent'].astype("int64")
df['company'] = df['company'].astype("int64")
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
df.info()
```

```python
for i in df['arrival_date_month']:
    df['arrival_date_month (int)'] = d.datetime.strptime(df['arrival_date_month'][0], '%B').month
```

```python
df2 = df[["arrival_date_year", "arrival_date_month (int)", "arrival_date_day_of_month"]].copy()
df2.columns = ["year", "month", "day"]
df['arrival date'] = pd.to_datetime(df2)
```
